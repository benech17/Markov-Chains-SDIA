## <a name="ex1">Exercise 1: Simulating $M/M/1/\infty$ and $M/M/1/K$ queues</a> [(&#8593;)](#content)

This exercise will focus on simulating an $M/M/1/\infty$ and an $M/M/1/K$ queue to illustrate some of the results covered in lecture 5. <!--in Chapter 2 and in TD.3-->

In the following, the parameter of the exponential distributions governing the inter-arrival and service times will be denoted $\beta$ (birth move) and $\delta$ (death move) respectively, with $0< \beta < \delta $. The time instants at which changes occur in the process $X(t)$ will be denoted by $T_n$ for $n \in \mathbb{N}$. We further define:

\begin{equation}
    \rho = \frac{\beta}{\delta}, \; X_n = X(T_n).
\end{equation}

### $M/M/1/\infty$-queue

1\. a) Using the lecture notes, implement a function to simulate a trajectory $X(t)$ of an $M/M/1/\infty$ process, where $X(t)$ represents the number of customers in the system at time $t$. 
<!--  from Chapter 2, Section 2.3.1 -->

> *Remark*: the signature of the function could be of the form:
``` python
def run_mm1inf(beta: float, delta: float, rng, niter: int):
    """
    Args:
        beta (float): birth rate.
        delta (float): death rate
        rng (int): random seed (or random generator).
        niter (int): number of changes (events) in 
                     the process.
    Raises:
        ValueError: error triggered if beta <= 0.
        ValueError: error triggered if delta <= 0.
    Returns:
        X (array_like): trajectory :math:`(X(t_n))`.
        T (array_like): time instants at which a change in 
                        the process occurs :math:`(t_n)`.
    """ 
```

````python
import numpy as np
import matplotlib.pyplot as plt
def run_mm1inf(beta,delta,rng,niter):
    """
    Simulate a continuous-time Markov chain representing an M/M/1 queue with infinite capacity.

    Parameters:
    - beta (float): The arrival rate (inverse of the mean inter-arrival time).
    - delta (float): The service rate (inverse of the mean service time).
    - rng (int): Seed for NumPy random number generator to ensure reproducibility.
    - niter (int): Number of iterations to simulate the Markov chain.

    Raises:
    - ValueError: If beta or delta are non-positive.

    Returns:
    - T (numpy.ndarray): Array of cumulative times at each iteration.
    - X (numpy.ndarray): Array of the queue length at each iteration.

    The function uses NumPy's random module to generate exponential random variables for the inter-arrival
    times and service times. It then simulates the M/M/1 queue process over 'niter' iterations and returns
    arrays 'T' and 'X' representing the cumulative times and queue lengths at each iteration, respectively.
    """
    if beta<=0 or delta<=0: # Check that beta and delta are positive
        raise ValueError("beta and delta must be positive")
    np.random.seed(rng) # Set the seed for the NumPy random number generator
    X=np.zeros(niter+1) # Initialize the queue length array with numpy array for efficiency
    T=np.zeros(niter+1)
    for i in range(niter):
        birth=np.random.exponential(1/beta) # Generate exponential random variables for inter-arrival times and service times
        death=np.random.exponential(1/delta)
        if X[i]==0: # If the queue is empty, the next event is an arrival necessarily
            X[i+1]=X[i]+1
            T[i+1]=T[i]+birth
        else:
            T[i+1]=T[i]+min(birth,death) # Otherwise, the next event is the minimum of an arrival and a departure
            if birth<death:
                X[i+1]=X[i]+1
            else:
                X[i+1]=X[i]-1
    return T,X
````

b) Display the generated trajectory of the process for $(\beta,\delta) = (5, 6)$ (e.g., use only the last 100 samples of the trajectory to obtain a representative illustration).

> *Hint*: use the function `step` from the matlplotlib library.

````python
beta=5
delta=6
rng=10
niter=1000
T,X=run_mm1inf(beta,delta,rng,niter)
plt.step(T[-100:],X[-100:])
plt.xlabel('Time')
plt.ylabel('Number of customers')
plt.title('Evolution of the number of customers in an M/M/1 queue (last 100 iterations)')
plt.show()
````

The values the queueing process takes seem coherent, if $beta<<delta$ the queueing process will tend to be empty and if $beta>>delta$ the queueing process will tend to increase indefinitely. Also, we do not stagnate in a state for a very long time.

2\. a) Display on the same graph the normalized histogram of $X(t)$ and it's stationary distribution $\pi$ defined by (see Proposition 2.1 in lecture 5)

\begin{equation}
    (\forall i \in \mathbb{N}), \; \pi(i) = (1 - \rho)\rho^i.
\end{equation}

What does this graph illustrate?

````python

K=int(max(X)) #maximum state reached
#we compute the time spent in each state
x_times=np.zeros(K+1)
for i in range(1,len(X)):
    x_times[int(X[i-1])]+=T[i]-T[i-1]
plt.bar(range(K+1),x_times/sum(x_times)) #we normalize the time spent in each state and print it
rho=beta/delta #compute rho, the intensity of the process
r=np.arange(0,max(X)+1) #the reached states
pi=np.array([rho**i for i in r]) #the stationary distribution
pi=pi/sum(pi) #we normalize the stationary distribution just in case
plt.scatter(r,pi)
plt.xlabel('Number of customers')
plt.ylabel('Probability')
plt.title('Stationary distribution of the number of customers in an M/M/1 queue')
plt.show()
````

This illustrate the mean time spent at each state. By increasing the number of iterations, we can see that the mean time spent at each state tends towards the stationary distribution when the number of iterations increases.

b) Empirically evaluate the average number of customers, whose value should be close close to the theoretical value $\frac{\rho}{1-\rho}$ (see Definition 2.3 p. 2 in lecture 5).

````python
#we compute the mean of the number of customers in the queue, pondering by the time spent in each state, and take the mean of this value over 400 simulations
N=400
X_mean=np.zeros(N)
X_means=np.zeros(N)
for i in range(N):
    T,X=run_mm1inf(beta,delta,i,niter)
    x_times=np.zeros(int(max(X))+1)
    for j in range(1,len(X)):
        x_times[int(X[j-1])]+=T[j]-T[j-1]
    X_mean[i]=np.dot(x_times,range(int(max(X))+1))/sum(x_times)
    X_means[i]=np.mean(X_mean[:i+1]) #we compute the mean of the mean values at each iteration
#plot the consecutive mean values
plt.plot(range(N),X_means-rho/(1-rho))
plt.xlabel('Simulation number')
plt.ylabel('Difference between empirical and theoretical mean')
plt.title('Convergence of the empirical mean to the theoretical mean')

print('Theoretical mean: ',rho/(1-rho))
print('Empirical mean final: ',np.mean(X_mean))
````

We can see that the mean number of customers ponderated by the time spent at each state tends towards the theoretical value when the number of iterations increases.

### $M/M/1/K$

Consider the case where the size of the capacity service $K$ is finite, i.e., where the system can only accommodate up to $K$ customers. In comparison with the previous case, any new customer trying to enter the system at a time instant $t$ will be systematically rejected if $X(t) = K$.

3\. a) Propose a variant of the function developed in 1.a) to simulate a trajectory of an $M/M/1/K$ process.

````python
def run_mm1K(beta,delta,rng,niter,K):
    """
    Simulate a continuous-time Markov chain representing an M/M/1 queue with finite capacity.

    Parameters:
    - beta (float): The arrival rate (inverse of the mean inter-arrival time).
    - delta (float): The service rate (inverse of the mean service time).
    - rng (int): Seed for NumPy random number generator to ensure reproducibility.
    - niter (int): Number of iterations to simulate the Markov chain.
    - K (int): The capacity of the queue.

    Raises:
    - ValueError: If beta or delta or K are non-positive.

    Returns:
    - T (numpy.ndarray): Array of cumulative times at each iteration.
    - X (numpy.ndarray): Array of the queue length at each iteration.

    The function uses NumPy's random module to generate exponential random variables for the inter-arrival
    times and service times. It then simulates the M/M/1 queue process with a finite capacity 'K' over 'niter'
    iterations and returns arrays 'T' and 'X' representing the cumulative times and queue lengths at each iteration,
    respectively. If the queue reaches its capacity, arrivals are blocked until there is space available in the queue.
    """
    if beta<=0 or delta<=0 or K<=0: # Check that beta, delta and K are positive
        raise ValueError("beta, delta and K must be positive")
    np.random.seed(rng) # Set the seed for the NumPy random number generator
    X=np.zeros(niter+1) # Initialize the queue length array with numpy array for efficiency
    T=np.zeros(niter+1)
    for i in range(niter):
        birth=np.random.exponential(1/beta) # Generate exponential random variables for inter-arrival times and service times
        death=np.random.exponential(1/delta)
        
        if X[i]==0: # If the queue is empty, the next event is an arrival necessarily
            X[i+1]=X[i]+1
            T[i+1]=T[i]+birth
        elif X[i]==K: # If the queue is full, the next event is a departure necessarily
            X[i+1]=X[i]-1
            T[i+1]=T[i]+death
        else:
            T[i+1]=T[i]+min(birth,death) # Otherwise, the next event is the minimum of an arrival and a departure
            if birth<death and X[i]<K:
                X[i+1]=X[i]+1
            else:
                X[i+1]=X[i]-1
    return T,X
````

b) Display the trajectory of the process for $(\beta,\delta, K) = (5, 6, 3)$.

````python
beta=5
delta=6
rng=10
niter=1000
K=3
T,X=run_mm1K(beta,delta,rng,niter,K)
plt.step(T[-100:],X[-100:])
#set y values_dict as [0,1,2,3]
plt.xlabel('Time')
plt.ylabel('Number of customers')
plt.title('Evolution of the number of customers in an M/M/1/K queue (last 100 iterations)')
plt.show()
````

We can see that the values stay between 0 and 3, which is coherent with the fact that the capacity of the system is 3 and the fact that we reject the customers when the system is full. The time to process the customers is higher than the time it tkaes a customer to arrive but as we have a capacity of 3, we most of the time stay under 3 customers which is also coherent.

b) Compute the theoretical average number of customers (prove its epxression), and estimate its value using the function developed in 3.a). <!--(see TD3, exercise 3)-->

Computing the theoretical average number of customers:

$$
    \mathbb{E}[X] = \sum_{i=0}^K i \pi(i) = \sum_{i=0}^K i \frac{(1-\rho)\rho^i}{1-\rho^{K+1}} = \frac{1-\rho}{1-\rho^{K+1}} \sum_{i=0}^K i \rho^i= \frac{1-\rho}{1-\rho^{K+1}} \rho \frac{d}{d\rho} \sum_{i=0}^K \rho^i = \frac{1-\rho}{1-\rho^{K+1}} \rho \frac{d}{d\rho} \frac{1-\rho^{K+1}}{1-\rho} .
$$
$$
    \mathbb{E}[X]= \frac{1}{1-\rho^{K+1}} \rho \frac{-(K+1)\rho^K(1-\rho)+1-\rho^{K+1}}{(1-\rho)} 
$$

````python
#we compute the mean of the number of customers in the queue, pondering by the time spent in each state, and take the mean of this value over 400 simulations
N=400
X_mean=np.zeros(N)
X_means=np.zeros(N)
for i in range(N):
    T,X=run_mm1K(beta,delta,i,niter,K)
   
    x_times=np.zeros(K+1)
    for j in range(1,len(X)):
        x_times[int(X[j-1])]+=T[j]-T[j-1]
    X_mean[i]=np.dot(x_times,range(K+1))/sum(x_times)
    X_means[i]=np.mean(X_mean[:i+1]) #we compute the mean of the mean values at each iteration
plt.plot(range(N),X_means-rho/(1-rho**(K+1))*rho*(1-(K+1)*rho**K*(1-rho)-rho**(K+1))/(1-rho))
plt.xlabel('Simulation number')
plt.ylabel('Difference between empirical and theoretical mean')
plt.title('Convergence of the empirical mean to the theoretical mean')
plt.show()

print('Theoretical mean: ',1/(1-rho**(K+1))*rho*(1-(K+1)*rho**K*(1-rho)-rho**(K+1))/(1-rho))
print('Empirical mean final: ',np.mean(X_mean))
````

We can see that the mean number of customers ponderated by the time spent at each state tends towards the theoretical value when the number of iterations increases once again.

````python

n=32
beta=0.6
niter=10000
mat=np.random.choice([-1,1],size=(n,n)) #we build a random matrix of size n*n with values -1 or 1
np.random.seed(2)
for k in range(niter):
    Y=mat.copy()
    i,j=np.random.randint(0,n,size=2) #select a random cell
    Y[i,j]=-Y[i,j] #flip the value of the cell
    delta_H = 2 * Y[i, j] * (Y[(i+1)%n, j] + Y[(i-1)%n, j] + Y[i, (j+1)%n] + Y[i, (j-1)%n]) #compute the change in energy

    if delta_H<0 or np.random.rand()<np.exp(-beta*delta_H): #if the change in energy is negative or if the probability is high enough, we accept the change
        mat[i,j]=-mat[i,j]

            
    if k%1000==0:
        plt.imshow(mat)
        plt.show()
````

We can see that the the image tends towards a state where each cell is different of every of its neighbours. The matrix tends to be alternating. It is to minimize the energy of the system. ( as it is the goal of the Ising model)

## <a name="ex3">Exercise 3: Simulated annealing for the *traveling salesman* problem</a> [(&#8593;)](#content)

This exercise will be focused on the implementation of a basic simulated annealing algorithm to minimize a function $f : E \subset \mathbb{N} \rightarrow \mathbb{R}$, where $E$ is finite. An application to the *traveling salesman* problem will then be considered to test the algorithm implementation. 

In general, a simulated annealing algorithm can be described as follows.

---
**Simulated annealing**

Set $x_0 \in E$, $T_0  > 0$. 

$n \leftarrow 0$

While $(n \leq N)$ and $(T_n > T_{\min})$

  1. Draw a point $y \sim Q(x_n, \cdot)$ in the neighborhood of $x_n$, and $u \sim \mathscr{U}([0,1])$ (where $\mathscr{U}$ is the uniform distribution)
  
  2. Compute the acceptance probability: 
  
$$
    p(f(x_n), f(y), T_n) = 
    \begin{cases}
        1 & \text{if } f(y) < f(x_n) \\
        e^{-(f(y) - f(x_n))/T_n} & \text{otherwise.}
    \end{cases}
$$
  
  3. Set $x_{n+1} = \begin{cases}
       y & \text{if } u \leq p(f(x_n), f(y), T_n) \\
       x_n & \text{otherwise.}
       \end{cases}$
  4. Set $T_{n+1} = \frac{T_0}{\log(n+2)}$
  5. $n \leftarrow n+1$

Return $x_N$, $\bigl( f(x_n) \bigr)_{1 \leq n \leq N}$

> *Note*: in practice, the transition kernel $Q$, the neighborhood of the current point $x_n$ needs to be defined by the user, depending on the problem of interest. The definition of the other elements will be specified later for the *traveling salesman* problem.

The algorithm implemented in 1. will be applied to the *traveling salesman* problem, briefly described in the following lines.

---
**The Traveling Salesman problem**

A salesman must find the shortest route going only once through $K$ cities, represented by the points $C_1, \dotsc, C_K$ in $\mathbb{R}^2$. In this case, the set of all possible routes $E$ contains $K!$ elements, which excludes an exhaustive search as $K$ increases (*e.g.*, for $K \geq 10$). For this application, the problem thus consists in finding a route, i.e., a permutation $\sigma$ in the symmetric group $\Sigma_K$, minimizing the function

\begin{equation}
    \sigma = \bigl( \sigma(1), \dotsc , \sigma(K) \bigr) \mapsto f(x) = \sum_{i = 1}^{K} \text{dist} (C_{\sigma(i)}, C_{\sigma(i+1)})
\end{equation}

with the convention $\sigma(K+1) = \sigma(1)$. In this case, simulated annealing randomly explores $\Sigma_K$ from one possible route to another located in its vicinity (which needs to be defined). 

In the following, the route $\tilde{\sigma} = \bigl( \tilde{\sigma}(1), \dotsc, \tilde{\sigma}(K) \bigr)$ will be said to be a neighbour of $\sigma = \bigl( \sigma(1), \dotsc, \sigma(K) \bigr)$ if there exists $1 \leq i < k \leq K $ such that:

\begin{equation}
   \bigl( \tilde{\sigma}(1), \dotsc ,\tilde{\sigma}(K) \bigr) = \bigl( \sigma(1), \dotsc, \sigma(i−1), \sigma(k), \sigma(k−1), \dotsc, \sigma(i+1), \sigma(i), \sigma(k+1), \dotsc, \sigma(K) \bigr). 
\end{equation}

For instance, for $K = 8$, the permutations $(1,2,3,4,8,7,6,5)$ and $(1,2,7,8,4,3,6,5)$ are neighbours (with $i = 3$, $k = 6$).
